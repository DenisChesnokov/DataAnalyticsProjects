```python
#Все названия пишутся с заглавной буквы, а типы населённых пунктов с маленькой.
#Выделим эти названия и проверим на дубли

clear_names = {} #словарь для выделенных имён в качестве ключей и списка названий типов.

for name in list(df['locality_name'].unique()): #перебираем названия из датасета

    clear_name = '' #в этой строке будем собирать названия населённого пункта из нескольких слов
    type_name = '' #тут будем собирать названия типов для каждого укороченного названия населённых пунктов 
    
    for word in name.split(): #перебираем слова в строке с названием из датасета        
        if word[0].isupper(): #если слово начинается с заглавной буквы добавляем к строке имени
            clear_name += word + ' '
        else:
            type_name += word + ' ' 
    
    #добавляем полученные значения в словарь. Если такого ключа еще нет то добавляем его и пустой список, 
    #затем плюсуем к списку новый элемент. Если такой ключ уже есть, то тоже добавляем к нему новый элемент
    clear_names[clear_name.strip()] = clear_names.get(clear_name.strip(), []) + ([type_name.strip()])

clear_names
```

Уберём типы поселений из названия

```python
#Выделим значения где в каждом значении списка есть слово посёлок
town_type = {}
for name, types in douple_type.items():
    type_len = len(types)
    for elem in types:
        if 'посёлок' in elem:
            type_len -= 1
    
    if type_len == 0:
        town_type[name] = types
        
town_type
```

```python
#Посмотрим оставшиеся значения
other_type = {}
for name, types in douple_type.items():
    if name not in contain_empty_type.keys() and name not in town_type.keys():
        other_type[name] = types
        
other_type
```

Поиск в сети показал, что:
    - деревня Фёдоровское и гп это одно и то же, заменим названия на посёлок
    - гп и пгт Павлово одно и то же, заменим названия на пгт, а вот село Павлово это другой населённый пункт
    - снт Рахья и пгт это разные населённые пункты, хотя находятся рядом
    - Зимитицы и деревня и посёлок находятся рядом, думаю статус поселения может влиять на цену, лучше оставить два разных названия
    - Калитино аналогично Зимитицам
    - Приладожский аналогично Зимитицам, думаю, что статус СНТ имеет существенное влияние на цену.
    - Щеглово аналогично Зимитицам
    - Рабитицы и деревня и посёлок находятся недалеко но не рядом, можно оставить два населённых пунтка
    - Гарболово деревня и посёлок это одно и тоже, заменим названия на посёлок

```python
#Напишем функцию замены значений
def change_city_type(full_city_name):

    #Уберём типы поселений из названий
    if 'Мурино' in full_city_name: 
        return 'Мурино'
    elif 'Кудрово'in full_city_name: 
        return 'Кудрово'
    elif 'Никольское' in full_city_name:
        return 'Никольское'
    elif 'Любань' in full_city_name:
        return 'Любань'

    #Приводим к единому типу гп и пгт
    
    elif 'Рощино' in full_city_name:
        return town_type['Рощино'][0] + ' Рощино'
    elif 'Советский' in full_city_name:
        return town_type['Советский'][0] + ' Советский'
    elif 'Большая Ижора' in full_city_name:
        return town_type['Большая Ижора'][1] + ' Большая Ижора'
    elif 'Лесогорский' in full_city_name:
        return town_type['Лесогорский'][1] + ' Лесогорский'
    elif 'Рябово' in full_city_name:
        return town_type['Рябово'][1] + ' Рябово'
    elif 'Мга' in full_city_name:
        return town_type['Мга'][1] + ' Мга'
    elif 'Вещево' in full_city_name:
        return town_type['Вещево'][1] + ' Вещево'
    elif 'Назия' in full_city_name:
        return town_type['Назия'][1] + ' Назия'
    elif 'Приветнинское' in full_city_name:
        return town_type['Приветнинское'][1] + ' Приветнинское'


    #Приводим оставшиеся значения
    
    elif 'Фёдоровское' in full_city_name:
        return other_type['Фёдоровское'][1] + ' Фёдоровское'
    elif 'Павлово' in full_city_name:
        if 'село' in full_city_name:
            return full_city_name
        else:
            return other_type['Павлово'][2] + ' Павлово'
    elif 'Гарболово' in full_city_name:
        return other_type['Гарболово'][1] + ' Гарболово'

    #в остальных случаях сохраняем название
    else:
        return full_city_name
```

```python
#Применим функцию
df['locality_name'] = df['locality_name'].apply(change_city_type)
```

```python
#Заполним пропуски для населенных пунктов, кроме Санкт-Петербурга

#Получаем список населённых пунктов
city_list = list(df.loc[(df['floors_total'].isna()) & (df['locality_name'] != 'Санкт-Петербург')]['locality_name'].unique())

#Создаем словарь медианной этажности
buildings_floor = {}

for city in city_list: #Записываем в словарь значение медианы по населённому пункту.
                            #приводим к типу int для будущей смены типа столбца
    buildings_floor[city] = [int(df.query('locality_name == @city')['floors_total'].median()), 
                             int(df.query('locality_name == @city')['floors_total'].max())]

buildings_floor
```

```python
#Создадим функцию заполнения пропусков в этажности дома
def set_buildings_floor(row):
    
    if row['floor'] > buildings_floor[row['locality_name']][0]: #проверяем что этаж квартиры больше медианы
        #если больше - устанавливаем максимум в этажность дома по группе
        row['floors_total'] = buildings_floor[row['locality_name']][1]
    
    else: #если меньше то приравниваем к медиане по группе
        row['floors_total'] = buildings_floor[row['locality_name']][0]
    return row
```

```python
#Создадим список с максимумом и медианой этажности в зависимости от расстояния до центра с шагом 5 км.

#Создаем словарь этажности
center_far = {}

for distance in range(5000, 35000, 5000): 
    center_far[distance] = [int(df.query('cityCenters_nearest > (@distance - 5000) \
                                    and cityCenters_nearest <= @distance \
                                    and locality_name == "Санкт-Петербург"')['floors_total'].median()),
                            
                            int(df.query('cityCenters_nearest > (@distance - 5000) \
                                    and cityCenters_nearest <= @distance \
                                    and locality_name == "Санкт-Петербург"')['floors_total'].max())]
   
center_far
```

```python
#Функция изменения этажности дома для Санкт-Петербурга
def set_buildings_floor_spb(row):
    for key in center_far.keys(): #чтобы не повторять одни и те же куски кода для каждой грппы создадим цикл
        if  row['cityCenters_nearest'] > (key - 5000) and row['cityCenters_nearest'] <= key: # проверяем попадание в гурппу
            if row['floor'] < center_far[key][0]: #проверяем, что этаж квартиры меньше медианы по группе
                row['floors_total'] = center_far[key][0] #если меньше, то этажность дома равна медиане по группе
            else:
                row['floors_total'] = center_far[key][1]    #иначе приравниваем этажность дома максимуму по группе
    return row
```

Исследуем этот вопрос при рассмотрении столбца числа комнат, но пока направшивается вывод об их удалении.

```python
#Расчитаем верхний и нижний пределы разброса значений для площади квартир в зависимотси от числа комнат
total_area_info.loc['lower_edge'] = total_area_info.loc['25%'] - (total_area_info.loc['75%'] - total_area_info.loc['25%']) * 1.5
total_area_info.loc['upper_edge'] = total_area_info.loc['75%'] + (total_area_info.loc['75%'] - total_area_info.loc['25%']) * 1.5
total_area_info
```

```python
#Изменим значения границ которые меньше минимального и больше максимального значения в данной выборке
def correct_edge(column):
    if column['lower_edge'] < column['min']:
        column['lower_edge'] = column['min']

    if column['upper_edge'] > column['max']:
        column['upper_edge'] = column['max']
    
    return column
```

```python
#Заменим значения границ
total_area_info.apply(correct_edge)
total_area_info
```

```python
#Проверим сколько квартир в каждой категории выходит за нижний предел.
for i in total_area_info.columns:
    print(f"Количество квартир с числом комнат {i} и площадью меньше нижнего предела:", \
          df.loc[(df['rooms'] == i) & (df['total_area'] < total_area_info.loc['lower_edge', i])]['total_area'].count()) 
```

```python
#Проверим сколько квартир в каждой категории выходит за верхний предел.
for i in total_area_info.columns:
    print(f"Количество квартир с числом комнат {i} и площадью больше верхнего предела:", \
          df.loc[(df['rooms'] == i) & (df['total_area'] > total_area_info.loc['upper_edge', i])]['total_area'].count()) 
```

Так как все значения выходящие за рамки пределов считаются аномальными выбросами, то от них лучше избавиться.

```python
for i in total_area_info.columns:
    #Удаляем значения меньше минимального значения размаха
    df = df.drop(df.loc[(df['rooms'] == i) & (df['total_area'] < total_area_info.loc['lower_edge', i])].index)
    
    #Удаляем значения выше максимальной границы размаха
    df = df.drop(df.loc[(df['rooms'] == i) & (df['total_area'] > total_area_info.loc['upper_edge', i])].index)
```
